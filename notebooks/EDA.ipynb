# Data Wrangling

Filling null values:

- **Orders** :
  - **order_approved_at** = 9/89316 null values. Missing values replaced with "Not Approved.
  - **order_delivered_timestamp** = 1889/89316 null values. Missing values replaced with "Not Delivered."
- **Products** :
  - **product_weight_g, product_length_cm, product_height_cm, product_width_cm** = 15/89316 null values. Missing values filled with their respective column means.
  - **product_category_name** = 308/89316 null values. Missing values filled with 'Unknown'






```python
# Filling null values for 'Orders' dataset
orders_df['order_approved_at'].fillna("Not Approved", inplace=True)  # Replace missing with "Not Approved"
orders_df['order_delivered_timestamp'].fillna("Not Delivered", inplace=True)  # Replace missing with "Not Delivered"

# Filling null values for 'Products' dataset
products_df['product_weight_g'].fillna(products_df['product_weight_g'].mean(), inplace=True)  # Fill with mean
products_df['product_length_cm'].fillna(products_df['product_length_cm'].mean(), inplace=True)
products_df['product_height_cm'].fillna(products_df['product_height_cm'].mean(), inplace=True)
products_df['product_width_cm'].fillna(products_df['product_width_cm'].mean(), inplace=True)

# Fill missing values in 'product_category_name' with 'Unknown'
products_df['product_category_name'].fillna('Unknown', inplace=True)

```

#Calculating Key Statistics


```python
# Merge datasets
merged_df = orders_df.merge(order_items_df, on='order_id', how='inner') \
                     .merge(customers_df, on='customer_id', how='inner') \
                     .merge(products_df, on='product_id', how='left')
```

```python
# Convert relevant columns to datetime
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'], errors='coerce')
merged_df['order_delivered_timestamp'] = pd.to_datetime(merged_df['order_delivered_timestamp'], errors='coerce')
merged_df['order_approved_at'] = pd.to_datetime(merged_df['order_approved_at'], errors='coerce')
merged_df['delivery_time_days'] = (merged_df['order_delivered_timestamp'] - merged_df['order_approved_at']).dt.days
```

```python
# Calculate Delivery Time
merged_df['delivery_time_days'] = (merged_df['order_delivered_timestamp'] - merged_df['order_purchase_timestamp']).dt.days
```

```python
# Calculate Key Statistics
# 1. Average Delivery Time
average_delivery_time = merged_df['delivery_time_days'].mean()
print(f"Average Delivery Time: {average_delivery_time:.2f} days")
```

```python
# 2. Order Frequency by Customer
order_frequency_by_customer = merged_df.groupby('customer_id').size().mean()
print(f"Average Order Frequency per Customer: {order_frequency_by_customer:.2f} orders")

```

```python
# Calculate Median Price and Quantity Ordered by Product Category
product_stats = merged_df.groupby('product_category_name').agg({
    'price': 'median',  # Median price
    'order_id': 'count'  # Count of orders as a proxy for quantity
}).rename(columns={'price': 'Median Price', 'order_id': 'Quantity Ordered'})

# Display the results
print(product_stats)
```

```python
# Add delivery time column
merged_df['delivery_time_days'] = (merged_df['order_delivered_timestamp'] - merged_df['order_approved_at']).dt.days
```

```python
# Calculate order frequency by date
order_frequency = merged_df.groupby(merged_df['order_purchase_timestamp'].dt.date).size()
```

```python
# Mapping Brazilian states to regions
state_to_region = {
    'AC': 'North', 'AL': 'Northeast', 'AP': 'North', 'AM': 'North', 'BA': 'Northeast',
    'CE': 'Northeast', 'DF': 'Central-West', 'ES': 'Southeast', 'GO': 'Central-West',
    'MA': 'Northeast', 'MT': 'Central-West', 'MS': 'Central-West', 'MG': 'Southeast',
    'PA': 'North', 'PB': 'Northeast', 'PR': 'South', 'PE': 'Northeast',
    'PI': 'Northeast', 'RJ': 'Southeast', 'RN': 'Northeast', 'RS': 'South',
    'RO': 'North', 'RR': 'North', 'SC': 'South', 'SP': 'Southeast',
    'SE': 'Northeast', 'TO': 'North'
}

# Add a region column to the customers dataset
customers_df['region'] = customers_df['customer_state'].map(state_to_region)
if 'customer_state' not in merged_df.columns:
    raise KeyError("'customer_state' column is missing in merged_df")
merged_df['region'] = merged_df['customer_state'].map(state_to_region)
delivery_status_by_region = merged_df.groupby(['order_year_month', 'region']).agg({
    'delivery_status': 'mean'
}).reset_index()

# Validate the mapping by checking unique regions
print(customers_df['region'].unique())
```

```python
# Ensure delivery status is binary (1 for success, 0 for failure)
merged_df['delivery_status'] = merged_df['order_delivered_timestamp'].notna().astype(int)
```

```python
# Extract year and month for seasonality analysis
merged_df['order_year_month'] = merged_df['order_purchase_timestamp'].dt.to_period('M')
```

```python
# Group by year-month and region to calculate delivery success rate
delivery_status_by_region = merged_df.groupby(['order_year_month', 'region']).agg({
    'delivery_status': 'mean'
}).reset_index()

```

# Exploratory Data Analysis

Delivery Time

```python
# Delivery time distribution visualization
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['delivery_time_days'].dropna(), bins=20, kde=True)
plt.title('Delivery Time Distribution (Days)', fontsize=14)
plt.xlabel('Delivery Time (Days)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Most deliveries are made within 010 days, with a peak around 2 days, but some take over 50 days, showing potential delays.

```python
# Visualization 4: Delivery Time by Product Category
plt.figure(figsize=(12, 8))
category_delivery_time = merged_df.groupby('product_category_name')['delivery_time_days'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=category_delivery_time.values, y=category_delivery_time.index, palette='coolwarm')
plt.title('Average Delivery Time by Product Category (Top 10)', fontsize=14)
plt.xlabel('Average Delivery Time (Days)', fontsize=12)
plt.ylabel('Product Category', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

This chart highlights that "diapers_and_hygiene" and "office_furniture" have the highest average delivery times, exceeding 20 days, suggesting possible delays in these categories.


```python
# 3. Order Date vs. Delivery Status
merged_df['delivery_status'] = merged_df['order_delivered_timestamp'].notna()
order_delivery_status = merged_df.groupby(merged_df['order_purchase_timestamp'].dt.date)['delivery_status'].mean()
plt.figure(figsize=(12, 8))
order_delivery_status.plot()
plt.title('Delivery Success Rate Over Time', fontsize=14)
plt.xlabel('Order Date', fontsize=12)
plt.ylabel('Delivery Success Rate', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```


The delivery success rate remains consistently near 100% for most periods, with occasional dips dropping as low as 80% in certain months in 2017, and a sharp decline to 0% during the initial period in late 2016 and the final period in 2018, likely due to early operational issues or incomplete data.

```python

# Delivery Time vs. Customer Location
plt.figure(figsize=(12, 8))
sns.boxplot(data=merged_df, x='customer_state', y='delivery_time_days')
plt.title('Delivery Time by Customer Location', fontsize=14)
plt.xlabel('Customer State', fontsize=12)
plt.ylabel('Delivery Time (Days)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Delivery times across most states cluster around 020 days, with So Paulo (SP) showing the shortest median times, while states like Acre (AC) and Roraima (RR) experience longer and more variable delivery times, with outliers exceeding 150 days.
